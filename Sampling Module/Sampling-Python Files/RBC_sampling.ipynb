# Red Blood Cell Sampling: Drugs

Set the environment

```python
import cobra.test
import pandas as pd
import numpy as np
import pandas as pd
import escher
from escher import Builder
from cobra.sampling import sample
import matplotlib.pyplot as plt
from cobrapy_bigg_client import client
pd.set_option('display.max_rows', 1000)
pd.set_option('display.width',1000)
pd.set_option('display.max_colwidth',None)
```

Load the red blood cell model iAB_RBC_283.json

```python
model = client.download_model('iAB_RBC_283', save=False) # Loading the model to the simulation
```

The summary of the model using default conditions

EX_co2_e, EX_fum_e, EX_lac__L_e, EX_mal__L_e, CO2t, EX_pyr_e, HEX1, HEX7, FUM, FUMtr, L_LACt2r, LDH_LG, MALt, GTHOrm, GTHPi, ME2, PGI, PYRt2, SBTD_D2, SBTR

```python
model.summary()
```

## Sampling the default model

```python
%%time
RBC_samples = sample(model, 5000, processes=8) #Set this to 1000 for faster, less accurate performance
RBC_samples.head(10) # Do not list all the samples, it could overload the browser memory
```

Using the reactions from the FVA example on the impact of drugs on the PGI reaction (RBC_FVA_Drugs.ipynb), let's look at the difference in flux distributions for the reactions whose flux span was changed by over 40%.
They include 'EX_co2_e', 'EX_fum_e', 'EX_lac__L_e', 'EX_mal__L_e', 'CO2t', 'EX_pyr_e', 'HEX1', 'HEX7', 'FUM', 'FUMtr', 'L_LACt2r', 'LDH_L', 'MALt', 'GTHOr', 'GTHPi', 'ME2', 'PGI', 'PYRt2', 'SBTD_D2', 'SBTR'.

```python
impacted_reactions = ['EX_co2_e', 'EX_fum_e', 'EX_lac__L_e', 'EX_mal__L_e', 'CO2t', 'EX_pyr_e', 'HEX1', 'HEX7', 
                      'FUM', 'FUMtr', 'L_LACt2r', 'LDH_L', 'MALt', 'GTHOr', 'GTHPi', 'ME2', 'PGI', 
                      'PYRt2', 'SBTD_D2', 'SBTR']

RBC_samples.hist(column = impacted_reactions, bins=100,figsize=(15,15), histtype = 'step');
```

## Knockout the PGI reaction

```python
model2 = client.download_model('iAB_RBC_283', save=False) # Loading the model to the simulation
model2.reactions.PGI.bounds = [0,0] # knock-out the PGI reaction

RBC_samples2 = sample(model2, 5000, processes=8) #Set this to 1000 for faster, less accurate performance
RBC_samples2.head(10) # Do not list all the samples, it could overload the browser memory
```

```python
#RBC_samples2.hist(column = ['CO2t', 'HEX1', 'HEX7', 'FUM', 'FUMtr', 'L_LACt2r', 'LDH_L', 'MALt', 'GTHOr', 
#                           'GTHPi', 'ME2', 'PGI', 'PYRt2', 'SBTD_D2', 'SBTR'], bins=100,figsize=(15,15),histtype = 'step');

RBC_samples2.hist(column = impacted_reactions, bins=100,figsize=(15,15), histtype = 'step');
```

## Plotting the difference between the default and diseased model (TPI knocked out)

Finding the impact to cell when the TPI reaction is set to zero (knocked-out). This corresponds to removing the TPI1 protein from the model. This corresponds to hemolytic anemia due to triosephosphate isomerase deficiency according to the OMIM database.

```python
reactions = ['CO2t', 'HEX1', 'HEX7', 'FUM', 'FUMtr', 'L_LACt2r', 'LDH_L', 'MALt', 'GTHOr', 
                           'GTHPi', 'ME2', 'PGI', 'PYRt2', 'SBTD_D2', 'SBTR']
fig, axes = plt.subplots(nrows=5, ncols=4)
j=0
for i in range(len(impacted_reactions)):
    normal_data=RBC_samples[impacted_reactions[i]]
    drug_data=RBC_samples2[impacted_reactions[i]]
    temp = pd.DataFrame({impacted_reactions[i]+ ': Normal data':normal_data, "Drug_data":drug_data})
    temp.plot.hist(bins=250, histtype = 'step',ax=axes[j%5,i%4],figsize=(15,15),density=True)
    if (i != 0 and (i+1)%4 == 0):
        j += 1
```
