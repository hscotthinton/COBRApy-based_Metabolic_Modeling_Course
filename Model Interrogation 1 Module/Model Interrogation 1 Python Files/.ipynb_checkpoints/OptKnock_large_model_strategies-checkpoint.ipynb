```python
reaction_ids = [r.id for r in model.reactions]

exclude = []
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Biomass and maintenance functions':
        exclude.append(reaction_ids[i])

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Extracellular exchange':
        exclude.append(reaction_ids[i])

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Transport, Outer Membrane':
        exclude.append(reaction_ids[i])

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Transport, Inner Membrane':
        exclude.append(reaction_ids[i])       

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Oxidative Phosphorylation':
        exclude.append(reaction_ids[i]) 
        
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Oxidative Phosphorylation':
        exclude.append(reaction_ids[i]) 
        
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Intracellular demand':
        exclude.append(reaction_ids[i]) 
        
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Transport, Outer Membrane Porin':
        exclude.append(reaction_ids[i])
        
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Cell Envelope Biosynthesis':
        exclude.append(reaction_ids[i])
        
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Murein Biosynthesis':
        exclude.append(reaction_ids[i])

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Glycerophospholipid Metabolism':
        exclude.append(reaction_ids[i])        

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Nucleotide Salvage Pathway':
        exclude.append(reaction_ids[i])         

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Murein Recycling':
        exclude.append(reaction_ids[i])         

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Membrane Lipid Metabolism':
        exclude.append(reaction_ids[i])          
        
for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Inorganic Ion Transport and Metabolism':
        exclude.append(reaction_ids[i])

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Lipopolysaccharide Biosynthesis / Recycling':
        exclude.append(reaction_ids[i])        

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'tRNA Charging':
        exclude.append(reaction_ids[i]) 

for i in range(len(reaction_ids)):
    if model.reactions.get_by_id(reaction_ids[i]).subsystem == 'Unassigned':
        exclude.append(reaction_ids[i])         

print(len(exclude))
#exclude
```

```python
from cameo.strain_design.deterministic import OptKnock
model = model_orig.copy()
#model.reactions.EX_o2_e.bounds = [0,0]
model.solver = 'gurobi'
optknock = OptKnock(model, fraction_of_optimum=0.01)
result2 = optknock.run(max_knockouts=2, target="EX_etoh_e", 
                       biomass="BIOMASS_Ec_iJO1366_core_53p95M",
                       #max_results = 10)
                       exclude_reactions = exclude,
                        )
result2
```
